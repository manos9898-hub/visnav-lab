```python
# ── Configuration ─────────────────────────────────────────────────────────────
# Set the session UUID here. All paths derive from this value.
SESSION_UUID = "REPLACE_WITH_SESSION_UUID"

import pathlib
WORKSPACES = pathlib.Path("/home/ubuntu/lab/workspaces")
SESSION_DIR = WORKSPACES / SESSION_UUID / f"session_{SESSION_UUID}"
META_FILE = SESSION_DIR / "session_meta.json"

assert SESSION_DIR.exists(), f"Session workspace not found: {SESSION_DIR}\nRun sync-sessions.sh first."
print(f"Session directory: {SESSION_DIR}")
```

```python
import json
with open(META_FILE) as f:
    meta = json.load(f)
print("session_meta.json loaded OK")
```

```python
print("── Identifiers ──────────────────────────────")
print(f"  session_id    : {meta['session_id']}")
print(f"  start_time    : {meta['start_time']}")
print(f"  end_time      : {meta['end_time']}")
print(f"  duration      : {meta['duration_seconds']:.1f} s")
print(f"  was_recovered : {meta['was_recovered']}")
if meta.get('recovery_reason'):
    print(f"  recovery_reason: {meta['recovery_reason']}")

print("\n── Provenance ───────────────────────────────")
print(f"  platform      : {meta['platform']}")
print(f"  os_version    : {meta['os_version']}")
print(f"  device_model  : {meta['device_model']}")
print(f"  app_name      : {meta['app_name']}")
print(f"  app_version   : {meta['app_version']}")
print(f"  schema_version: {meta['schema_version']}")
```

```python
print("── Camera ───────────────────────────────────")
print(f"  resolution      : {meta['capture_resolution']}")
print(f"  fps_target      : {meta['capture_fps_target']}")
print(f"  jpeg_quality    : {meta['capture_jpeg_quality']}")
print(f"  stabilization   : {meta['stabilization_state']}")
print(f"  focal_length_mm : {meta.get('camera_focal_length_mm')}")
print(f"  vertical_fov_deg: {meta.get('camera_vertical_fov_deg')}")

print("\n── Audio ────────────────────────────────────")
print(f"  codec           : {meta['audio_codec']}")
print(f"  bitrate_kbps    : {meta['audio_bitrate_kbps']}")
print(f"  sample_rate_hz  : {meta['audio_sample_rate_hz']}")
print(f"  duration_seconds: {meta['audio_duration_seconds']:.1f}")
print(f"  input_device    : {meta['audio_input_device_type']}")
```

```python
print("── Sensors Present ──────────────────────────")
for s in meta.get('sensors_present', []):
    print(f"  {s['sensor_type']:30s} {s.get('name','')}")

print("\n── Sensors Absent ───────────────────────────")
for s in meta.get('sensors_absent', []):
    print(f"  {s}")
```

```python
print("── Timestamp Sources ────────────────────────")
print(f"  camera_ts_source      : {meta.get('camera_ts_source')}")
print(f"  imu_ts_source         : {meta.get('imu_ts_source')}")
print(f"  gps_ts_source         : {meta.get('gps_ts_source')}")
print(f"  cross_sensor_alignment: {meta.get('cross_sensor_alignment')}")
```

```python
health = meta.get('health_summary', {})
print("── Health Summary ───────────────────────────")
print(f"  camera_obscuration_events : {len(health.get('camera_obscuration_events', []))}")
print(f"  mic_offline_events        : {len(health.get('mic_offline_events', []))}")
print(f"  sensor_dropout_events     : {len(health.get('sensor_dropout_events', []))}")
print(f"  timestamp_coherence_fails : {health.get('timestamp_coherence_failures', 0)}")
print(f"  config_parse_failure      : {health.get('config_parse_failure', False)}")

if health.get('sensor_dropout_events'):
    print("\n  Dropout detail:")
    for d in health['sensor_dropout_events']:
        print(f"    {d['sensor_type']} dropped at ts_app {d['ts_app_dropout']}")
```

```python
if meta['platform'] == 'ios' and 'arkit_version' in meta:
    print("── ARKit ────────────────────────────────────")
    print(f"  arkit_version            : {meta['arkit_version']}")
    print(f"  world_tracking_config    : {meta['world_tracking_configuration']}")
    print(f"  session_origin_gps       : {meta.get('session_origin_gps')}")
else:
    print("No ARKit data (Android session or ARKit not used)")
```
